```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import pairwise_distances
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from scipy.sparse import csr_matrix
import networkx as nx
import numpy as np
from collections import defaultdict
from community import community_louvain
from collections import Counter, defaultdict


# Function to reduce diagnosis categories by grouping similar codes
def group_diagnoses(df, diag_cols):
    for col in diag_cols:
        # Keep the first two digits to group by ICD9 hierarchy
        df[col] = df[col].apply(lambda x: str(x)[:1] if pd.notna(x) else 'Other')
    return df

# Apply grouping
diag_cols = ['diag_1', 'diag_2','diag_3']
sample = group_diagnoses(sample, diag_cols)

# Filter for frequent categories (frequency-based encoding)
def encode_frequent_categories(df, column, threshold=20):
    counts = df[column].value_counts()
    frequent = counts.index[:threshold]
    df[column] = df[column].apply(lambda x: x if x in frequent else 'Other')
    return df

for col in diag_cols:
    sample = encode_frequent_categories(sample, col)

# One-hot encoding
categorical_cols = diag_cols + ['change', 'diabetesMed', 'readmitted']  + [
        'metformin', 'repaglinide', 'nateglinide', 'glimepiride', 'glipizide', 
        'glyburide', 'tolbutamide', 'pioglitazone', 'rosiglitazone', 
        'acarbose', 'miglitol', 'tolazamide', 'insulin'
    ]
one_hot_encoder = OneHotEncoder()
encoded_categorical = one_hot_encoder.fit_transform(sample[categorical_cols]).toarray()

# Standardize the numerical features
numerical_cols = [
    'time_in_hospital', 'num_lab_procedures', 'num_procedures', 
    'num_medications', 'number_outpatient', 'number_emergency', 
    'number_inpatient', 'number_diagnoses'
]
scaler = StandardScaler()
scaled_numerical = scaler.fit_transform(sample[numerical_cols])

# Concatenate standardized numerical and encoded categorical features
processed_data = np.hstack([scaled_numerical, encoded_categorical])


# Compute pairwise Euclidean distances and modularity
distances = pairwise_distances(processed_data, metric='euclidean')
sigma = np.std(distances)
W = np.exp(-distances**2 / (2 * sigma**2))
threshold = 0.1
W[W < threshold] = 0
W_sparse = csr_matrix(W)

# Build graph and calculate modularity
G = nx.from_scipy_sparse_array(W_sparse)

# Use the Louvain algorithm to find communities
partition = community_louvain.best_partition(G)
communities = [[node for node in partition if partition[node] == comm] for comm in set(partition.values())]

# Calculate modularity score
modularity_score = nx.community.modularity(G, communities)
print(f"Modularity score with grouped diagnoses: {modularity_score}")

# Count the size of each community
community_sizes = Counter(partition.values())

# Initialize dictionary to store readmission counts by community
community_readmission_counts = defaultdict(Counter)

# Count readmission values for nodes in each community with size > 100
for node, community in partition.items():
    if community_sizes[community] > 10:
        readmitted_value = sample.iloc[node]['readmitted']  # Ensure index matches
        community_readmission_counts[community][readmitted_value] += 1

# Display counts for communities with more than 100 nodes
for community, counts in community_readmission_counts.items():
    print(f"Community {community} (Size: {community_sizes[community]}): {dict(counts)}")
```

```python
# Identify isolated nodes
isolated_nodes = list(nx.isolates(G))
print(f"Isolated nodes found: {isolated_nodes}")

# Remove isolated nodes from the graph
G.remove_nodes_from(isolated_nodes)
```

```python
import numpy as np
import networkx as nx
from sklearn.metrics import pairwise_distances
from community import community_louvain
from deap import base, creator, tools, algorithms
from sklearn.preprocessing import StandardScaler, OneHotEncoder


# GA Parameters
POP_SIZE = 20     # Population size
GENS = 10         # Number of generations
MUT_PROB = 0.2    # Mutation probability
CX_PROB = 0.5     # Crossover probability
K_BEST = 10       # Number of top features


# Fitness function to evaluate each individual's performance
def evaluate_individual(individual):
    # Select features based on binary chromosome
    selected_features = processed_data[:, np.array(individual) == 1]
    
    # Build the graph based on selected features
    distances = pairwise_distances(selected_features, metric='euclidean')
    sigma = np.std(distances)
    W = np.exp(-distances**2 / (2 * sigma**2))
    W[W < 0.1] = 0  # Threshold for sparsity
    G = nx.from_numpy_array(W)
    
    # Community detection and modularity score as fitness
    partition = community_louvain.best_partition(G)
    communities = [{node for node, comm in partition.items() if comm == c} for c in set(partition.values())]
    modularity = nx.community.modularity(G, communities)
    
    return (modularity,)

# GA setup with DEAP
def setup_genetic_algorithm():
    creator.create("FitnessMax", base.Fitness, weights=(1.0,))  # Maximizing modularity
    creator.create("Individual", list, fitness=creator.FitnessMax)
    
    toolbox = base.Toolbox()
    toolbox.register("attr_bool", np.random.randint, 0, 2)  # Binary gene, 0 or 1
    toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, processed_data.shape[1])
    toolbox.register("population", tools.initRepeat, list, toolbox.individual)
    
    toolbox.register("evaluate", evaluate_individual)
    toolbox.register("mate", tools.cxUniform, indpb=0.5)
    toolbox.register("mutate", tools.mutFlipBit, indpb=0.1)
    toolbox.register("select", tools.selTournament, tournsize=3)
    
    return toolbox

# Run GA
def run_genetic_algorithm():
    toolbox = setup_genetic_algorithm()
    pop = toolbox.population(n=POP_SIZE)
    
    algorithms.eaSimple(pop, toolbox, cxpb=CX_PROB, mutpb=MUT_PROB, ngen=GENS, verbose=True)
    
    best_individual = tools.selBest(pop, k=1)[0]
    best_features_indices = [i for i, val in enumerate(best_individual) if val == 1]
    
    return best_individual, best_features_indices

# Run the full process
best_individual, best_features_indices = run_genetic_algorithm()

print("Best feature subset (selected features indices):", best_features_indices)
print("Best individual modularity score:", evaluate_individual(best_individual)[0])
```
